```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
ds1radiotemp = pd.read_csv("../../dataset_1_radioTemperatures_20210303.csv",sep=';',dtype={'id_audit':np.str})
ds2powerclass = pd.read_csv("../../dataset_2_powerClass_20210303.csv", sep=";", dtype={'id_audit':np.str})
ds3radioid = pd.read_csv("../../dataset_3_radioId_20210303.csv",sep=";", dtype={'id_audit':np.str})
```

```python
ds1radiotemp.dropna(inplace=True)
```

```python
sensorData = ds1radiotemp.pivot_table(index=['id_audit'], columns=['field'],values=['value'])
sensorData = sensorData.droplevel(0, axis=1).reset_index()
sensorData.id_audit = sensorData.id_audit.astype(str)
```

```python
sensorData.drop(['TSensor_C1_','TSensor_D11_','TSensor_D12_','TSensor_D13_','TSensor_D14_','TSensor_D15_'
                ,'TSensor_D16_','TSensor_D17_','TSensor_D2_','TSensor_D3_','TSensor_D7_','TSensor_P39_'
                ,'TSensor_P40_','TSensor_P41_','TSensor_P42_','TSensor_P43_','TSensor_P44_','TSensor_P45_'
                ,'TSensor_P46_','TSensor_T63_','TSensor_T64_','TSensor_W73_','TSensor_W74_'], axis=1, inplace=True)
```

```python
sensorDataRed = sensorData.iloc[:,1:].applymap(lambda x: np.nan if (x < -40 or x> 125) else x)
sensorDataRed['id_audit'] = sensorData.id_audit.values
sensorDataRed.dropna(inplace=True)
sensorDataRed.info(verbose=True,null_counts=True)
```

```python
sensorDataRed.head()
```

```python
sensors = sensorDataRed.columns.values[:-1]
sensorDataRed.shape,sensors
```

```python
#from sklearn.cluster import DBSCAN
#dbscan=DBSCAN()
from sklearn.ensemble import IsolationForest
dbscan = IsolationForest(contamination=0.01)
```

```python
masks = []
for aSensor in sensors:
    yhat = dbscan.fit_predict(sensorDataRed[aSensor].values.reshape(-1, 1))
    mask = yhat == -1
    masks.append(mask)
```

```python
import pickle
```

```python
with open("masksData.dat", "wb") as fp:
    pickle.dump(masks,fp)
```

```python
corr_matrix = sensorDataMasked.iloc[:,:-1].corr()
#the matrix is symmetric so we need to extract upper triangle matrix without diagonal (k = 1)
sol = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
                  .stack()
                  .sort_values(ascending=False))
```

```python
corr_matrix.describe()
```

```python
sollist = sol.groupby(level=0).head(1)
sollist.tail()
```

```python
sollist.xs('TSensor_T58_', level=1, drop_level=False)
```

```python
sollist.index.get_level_values(0)
```

```python
from sklearn.linear_model import LinearRegression
lnclf1 = LinearRegression()
lnclf2 = LinearRegression()
```

```python
lnclf1.fit(sensorDataMasked['TSensor_D19_'].values.reshape(-1, 1), sensorDataMasked['TSensor_R54_'].values.reshape(-1, 1))
```

```python
ypred1 = lnclf1.predict(sensorDataRed['TSensor_D19_'].values.reshape(-1, 1))
```

```python
residuals1 = sensorDataRed['TSensor_R54_'].values.reshape(-1, 1)-ypred1
```

```python
plt.plot(sensorDataRed['TSensor_D19_'].values.reshape(-1, 1),residuals1, 'o', color='darkblue')
plt.title("Residual Plot")
plt.xlabel("Last sensor")
plt.ylabel("Residual")
```

```python
lnclf2.fit(sensorDataMasked['TSensor_R54_'].values.reshape(-1, 1), sensorDataMasked['TSensor_D19_'].values.reshape(-1, 1))
ypred1 = lnclf1.predict(sensorDataRed['TSensor_R54_'].values.reshape(-1, 1))
residuals1 = sensorDataRed['TSensor_D19_'].values.reshape(-1, 1)-ypred1
plt.plot(sensorDataRed['TSensor_R54_'].values.reshape(-1, 1),residuals1, 'o', color='darkblue')
plt.title("Residual Plot")
plt.xlabel("Last sensor")
plt.ylabel("Residual")
```

```python
fig, ax = plt.subplots(figsize=(20,10)) 
ax = plt.axes(yscale='linear')
ax.grid();
sns.residplot(x='TSensor_D19_', y='TSensor_R54_', data=sensorDataRed, scatter_kws=dict(s=30), ax=ax)
```

```python
residuals1.shape, residuals1.mean(),residuals1.std(),abs(residuals1.mean()) + 3 * residuals1.std()
```

```python
resMask1 = np.abs(residuals1) > (residuals1.mean() + 3 * residuals1.std())
```

```python
resMask1.shape, maskpd.shape, type(resMask1)
```

```python
finalMasks = pd.DataFrame(index=sensors, columns=maskpd.columns)
finalMasks.loc['TSensor_D19_']= resMask1.reshape(1,-1)
```

```python
finalMasks.loc['TSensor_D19_'].value_counts()
```

```python
sens19_54 = sensorDataRed[['TSensor_D19_','TSensor_R54_']].copy()
sens19_54['resMask'] =finalMasks.loc['TSensor_R54_'].values.reshape(-1,1)
```

```python
fig, ax = plt.subplots(figsize=(20,10)) 
ax = plt.axes(yscale='linear')
ax.grid();
sns.scatterplot(data = sens19_54, x = "TSensor_D19_", y = "TSensor_R54_", hue="resMask",ax=ax, palette="deep")
```

```python
finalMasks = pd.DataFrame(index=sensors, columns=maskpd.columns)
done = []
for index, value in (sollist.items()):
    #print(index[0], index[1],value, done)
    if index[0] not in done:
        print("Processing as first ... " + index[0] + " refer to " + index[1])
        lnclf1.fit(sensorDataMasked[index[1]].values.reshape(-1, 1), sensorDataMasked[index[0]].values.reshape(-1, 1))
        ypred1 = lnclf1.predict(sensorDataRed[index[1]].values.reshape(-1, 1))
        residuals1 = sensorDataRed[index[0]].values.reshape(-1, 1)-ypred1
        resMask1 = np.abs(residuals1) > (abs(residuals1.mean()) + 5 * residuals1.std())
        finalMasks.loc[index[0]] = resMask1.reshape(1,-1)
        done.append(index[0])
    
    if index[1] not in done:
        print("Processing as second ... " + index[1] + " refer to " + index[0])
        lnclf2.fit(sensorDataMasked[index[0]].values.reshape(-1, 1), sensorDataMasked[index[1]].values.reshape(-1, 1))
        ypred2 = lnclf1.predict(sensorDataRed[index[0]].values.reshape(-1, 1))
        residuals2 = sensorDataRed[index[1]].values.reshape(-1, 1)-ypred2
        resMask2 = np.abs(residuals2) > residuals2.mean() + 5 * residuals2.std()
        finalMasks.loc[index[1]] = resMask2.reshape(1,-1)
        done.append(index[1])
      
```

```python
sens19_54 = sensorDataRed[['TSensor_D19_','TSensor_R54_']].copy()
sens19_54['resMask'] = finalMasks.loc['TSensor_R54_'].values.reshape(-1,1) | finalMasks.loc['TSensor_D19_'].values.reshape(-1,1)
```
